2. We have calculated the two-body density matrix for bcs_s using the following formula:

$$
\rho_{\uparrow \downarrow}^{(2)}(r)
= \frac{
\int |\Psi(\mathbf{R})|^2\,
\frac{\Psi(\mathbf{r}_1+\mathbf{r}',\,\mathbf{r}_2+\mathbf{r}')}{\Psi(\mathbf{r}_1,\mathbf{r}_2)}
\,\delta\!\left(|\mathbf{r}'|-r\right)\,d\mathbf{R}\,d\mathbf{r}'
}{
\int |\Psi(\mathbf{R})|^2\,d\mathbf{R}
},~~~~  C(r) = \frac{\rho_{\alpha\beta}^{(2)}(r)}{\rho_{\alpha\beta}^{(2)}(0)}
$$

It can be see that for both $N_e = 8$ and $N_e = 16$, $C(r)$ of psiformer and ferminet is very close, with $N_e =16$ psiformer slightly closer to the target wavefunction. Although both networks captures the plateaus at large $r$, the structures of $C(r)$ are still differs from the target wavefunctions.

4. We have adopted a simple slater jastrow ansatz and optimized its variational parameters to match laughlin_m3. The slater jastrow ansatz has the following form:

$$
\Psi
=
\det\!\left[
\sum_{n=1}^{N}
{C}_{jn}\,
z_i^{\,n}\,
e^{-\frac{|r_i|^2}{4l_B^2}}
\right]
\;
\exp\!\left[
\alpha \sum_{i<j} |\mathbf{r}_i - \mathbf{r}_j|^2
\right].
$$

where ${C}_{jn}$ and $\alpha$ are variational parameters. It can be seen that at $N_e=2$, the Slater-Jastrow ansatz can capture partially the phase structure. However, at $N_e=3$, it fails to represent the nodal structure of the target laughlin_m3 wavefunction.

6. We have included one extra target wavefunction for Fermi liquid. The wavefunction uses plane-wave orbitals with a Mc Millan correlator:

$$
\Psi(\mathbf r_1, \mathbf r_2, \ldots)
=
\det\!\big[\phi_j(\mathbf r_i)\big]\,
\exp\!\left[
-\sum_{i<j}
\left(\frac{b_0}{r_{ij}}\right)^m
\right],
\qquad
\phi_j(\mathbf r)
=
\begin{cases}
\cos(\mathbf k_j\!\cdot\!\mathbf r),\\
\sin(\mathbf k_j\!\cdot\!\mathbf r),
\end{cases}
$$

We choose $b_0 = 1$ and $m=4$. To keep the wavefunction close shell, we choose $N_e$ such that all orbitals within $\mathbf k < |k_{\rm cut}|^2$ is included. As can be seen from below, the network can reach $F=0.94$ at $N_e=13$. This places the difficulty of Fermi liquid lower than topological wavefunctions and superconducting wavefunctions, but higher than Wigner crystal.
